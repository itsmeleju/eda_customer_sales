```python
#Q1 - most used payment method (and least use)

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('C:/Datascience/Assignment/KAGGLE/cust_time/df.xlsx')

# Assume the payment method column is named 'payment_method'
payment_method_counts = df['Payment'].value_counts()

# Print the top 5 most used payment methods
print(payment_method_counts) #.nlargest(5)

# Plot a bar chart to visualize the results
plt.figure(figsize=(8, 6))
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%')
plt.title('Most Used Payment Methods')
plt.xlabel('Payment Method')
plt.show('payment_method_counts')

# Assume the column name is 'column_name'
#Max used payment

# Get the most used payment method
most_used_payment = payment_method_counts.index[0]
most_used_payment_count = df['Payment'].value_counts().max()
print(f"The most used payment method is {most_used_payment} with a count of {most_used_payment_count}.")

#most_payment_name = kfc['Payment'].value_count().max()
#print("The most popular value is:", most_payment_name)


#least used payment

# Get the most used payment method
least_used_payment = payment_method_counts.index[-1]
least_payment_value = df['Payment'].value_counts().min()
print(f"The least used payment method is {least_used_payment} with a count of {least_payment_value}.")

#least_payment_value = df['Payment'].min()
#print("The least popular value is:", least_payment_value)
```

```python
#4 age category of the customers
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
df = pd.read_excel('C:/Datascience/Assignment/KAGGLE/cust_time/df.xlsx')
# Assume the date of birth column is named 'Date of Birth'
df['Birthdate'] = pd.to_datetime(df['Birthdate'])

# Calculate the age
df['Age'] = df['Birthdate'].apply(lambda x: (date.today() - x.date()).days // 365)

# Alternatively, you can use the following code to calculate the age
def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['Age'] = df['Birthdate'].dt.date.apply(calculate_age)


#4 - age category of the customers

# Assume the age column is named 'Age'
age_bins = [18, 25, 35, 45, 55, 65]
age_labels = ['15-24', '25-34', '35-44', '45-80', '55-80']

# Create age categories using pd.cut
age_categories = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Get the count of each age category
age_category_counts = age_categories.value_counts()

# Print the count of each age category
print(age_category_counts)

# Plot a bar chart to visualize the results
plt.figure(figsize=(8, 6))
plt.bar(age_category_counts.index, age_category_counts.values)
plt.title('Age Categories of Customers')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.show()

```

```python
#9 most visited month and least visited month
import pandas as pd
import datetime as date
#import matplotlib.pyplot as plt
df = pd.read_excel('C:/Datascience/Assignment/KAGGLE/cust_time/df.xlsx')
dates = ['Date']  # list of dates in the format YYYY-MM-DD
def extract_month(date):
    """Extract month from datetime object"""
    return date.month
df['Date'] = pd.to_datetime(df['Date'])
#Extract months
months = df['Date'].apply(extract_month)
# Count occurrences of each month
month_counts = months.value_counts()
#           """Adding month info to code"""
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
#name  most_least_branch = Branch_visted.index[-1]
most_visited_month = month_counts.idxmax()
least_visited_month = month_counts.idxmin()

print(f"Most visited month: {month_names[most_visited_month]}")
print(f"Least visited month: {month_names[least_visited_month]}")

```
